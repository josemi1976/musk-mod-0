1. Crea un array de enteros 4X2 e imprime sus atributos. Nota: El elemento debe ser de tipo unsigned int16. Imprime los siguientes atributos: 


- La shape del array.
- Las dimensiones del array.
- El tamaño de cada elemento del array en bytes.

````python
import numpy as np

firstArray = np.empty([4,2], dtype = np.uint16) 
print(firstArray)

print("Attributes")
print("Shape: ", firstArray.shape)
print("Dimensiones: ", firstArray.ndim)
print("Tamaño en bytes de cada elemento: ", firstArray.itemsize)
````

5. Crea una matriz de resultados sumando las siguientes dos matrices de NumPy. A continuación, modifica la matriz de resultados calculando el cuadrado de cada elemento.

```
arrayOne = numpy.array([[5, 6, 9], [21 ,18, 27]])
arrayTwo = numpy.array([[15 ,33, 24], [4 ,7, 1]])
```

````python
arrayOne = np.array([[5, 6, 9], [21 ,18, 27]])
arrayTwo = np.array([[15 ,33, 24], [4 ,7, 1]])

resultArray  = arrayOne + arrayTwo
print(resultArray)

for num in np.nditer(resultArray, op_flags = ['readwrite']):
    num[...] = num*num
print(resultArray)
````

7. Ordena el siguiente array de NumPy:
    
- Caso 1: Ordenar el array por la segunda fila
- Caso 2: Ordenar el array por la segunda columna
    
```
sampleArray = numpy.array([[34,43,73],[82,22,12],[53,94,66]])
```

````python
sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]]) 
print(sampleArray)

sortArrayByRow = sampleArray[:,sampleArray[1,:].argsort()]
print(sortArrayByRow)

sortArrayByColumn = sampleArray[sampleArray[:,1].argsort()]
print(sortArrayByColumn)
````

8. Imprime el máximo del eje 0 y el mínimo del eje 1 de la siguiente matriz bidimensional:

```
sampleArray = numpy.array([[34,43,73],[82,22,12],[53,94,66]])

```


````python
sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]]) 
print (sampleArray)

minOfAxisOne = np.amin(sampleArray, 1) 
print(minOfAxisOne)

maxOfAxisOne = np.amax(sampleArray, 0) 
print(maxOfAxisOne)
````

9. Elimina la segunda columna de una matriz dada e inserta la siguiente columna nueva en su lugar.

```
sampleArray = numpy.array([[34,43,73],[82,22,12],[53,94,66]])
```
